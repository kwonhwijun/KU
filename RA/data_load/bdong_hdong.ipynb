```python
from sqlalchemy import create_engine
from sqlalchemy.sql import text
query =  '''
CREATE TABLE conn_code_now AS
SELECT 법정동코드, MAX(CAST(행정동코드 AS INTEGER)) AS 행정동코드 FROM conn_code WHERE 말소일자 = ""
GROUP BY 법정동코드
'''
query = '''
SELECT COUNT(*) FROM conn_code_now
'''
eng = create_engine("sqlite:///real_estate.db")

with eng.connect() as conn:
    result = conn.execute(text(query))
    for row in result :
        print(row)
```
