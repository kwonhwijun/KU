```python
from sqlalchemy import create_engine
from sqlalchemy.sql import text
```

```python
eng = create_engine("sqlite:///data_load/real_estate.db")
query = "SELECT name FROM sqlite_master WHERE type = 'table'"

all_table =[]
with eng.connect() as conn :
    tables = conn.execute(text(query))
    for table_name in tables :
        all_table.append(table_name)
        result = conn.execute(text(f"SELECT COUNT(*) FROM {table_name[0]}"))
        print(f"테이블명 :{table_name[0]} 데이터의 수 :{result.scalar()}")
    
```

```python
# 필요한 테이블만 남기기

my_table = ['apt', 'apt_distinct_temp2', 'apt_distinct_coded_temp', 'conn_code']
table = [j for j in [i[0] for i in all_table] if j not in  my_table]

with eng.connect() as conn :
    for table_name in table :
        query = f"DROP TABLE IF EXISTS {table_name}"
        conn.execute(text(query))

```

```python
# 엑셀로 옮기기
import pandas as pd
import os
engine = create_engine("sqlite:///data_load/real_estate.db")

base_url = '/Users/hj/Dropbox/real_estate/data/apt'

for year in range(2006, 2025):
    # 연도별 폴더 경로 생성하기
    year_dir = os.path.join(base_url, str(year))
    # 폴더 없으면 폴더 생성
    if not os.path.exists(year_dir):
        os.makedirs(year_dir)
    # 저장
    for month in range(1,13):
        df = pd.read_sql(f'SELECT * FROM apt_distinct_coded_temp WHERE 년 = {year} AND 월 = {month}', con=engine)
        file_path = os.path.join(year_dir, f"아파트_매매_{year}년_{month:02}월.csv")
        df.to_csv(file_path, index=False)
    print(f"{year}년 저장완료")
```

```python
import os
import shutil

# 상위 폴더 경로 설정
base_url = '/Users/hj/Dropbox/real_estate/data/apt'

for year in range(2006, 2025):
    year_dir = os.path.join(base_url, str(year))
    if os.path.exists(year_dir):
        # 해당 연도 폴더 내의 모든 파일에 대해 반복
        for filename in os.listdir(year_dir):
                # 파일명 분석
                parts = filename.split('_')
                # 월 부분을 두 자리 숫자로 포매팅
                year_str = parts[2].replace('년', '')
                month = parts[3].replace(".xlsx", "")
                month_str = parts[3].replace('월.xlsx', '')
                if month_str.isdigit():
                    new_month_str = f"{int(month_str):02}월.xlsx"  # 두 자리 숫자 포매팅
                    new_filename = f"{parts[0]}_{parts[1]}_{year_str}년_{new_month_str}"
                    # 파일명 변경
                    old_file_path = os.path.join(year_dir, filename)
                    new_file_path = os.path.join(year_dir, new_filename)
                    shutil.move(old_file_path, new_file_path)
                    print(f"Renamed '{filename}' to '{new_filename}'")
                print(str(month).replace("월", ""))
```
